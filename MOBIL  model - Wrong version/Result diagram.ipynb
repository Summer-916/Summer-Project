```python
import json
import matplotlib.pyplot as plt

def plot_single_real_value(file_path, index):
    with open(file_path, 'r') as f:
        real_data = [json.loads(line.strip()) for line in f]

    if index < len(real_data):
        entry = real_data[index]
        timestamps = entry['timestamps']
        speeds = entry['vehicle_speeds']
        accelerations = entry['vehicle_accelerations']

        # 
        relative_timestamps = [ts - timestamps[0] for ts in timestamps]

        # t-v
        plt.figure(figsize=(10, 5))
        plt.plot(relative_timestamps, speeds, label="Empirical")
        plt.xlabel('Time (s)')
        plt.ylabel('Speed (mph)')
        plt.title(f'Time vs Speed for Vehicle')
        plt.legend()
        plt.grid(True)
        plt.show()

        # t-a
        plt.figure(figsize=(10, 5))
        plt.plot(relative_timestamps, accelerations, label="Empirical")
        plt.xlabel('Time (s)')
        plt.ylabel('Acceleration (mph/s)')
        plt.title(f'Time vs Acceleration')
        plt.legend()
        plt.grid(True)
        plt.show()

# 
file_path = 'trajectories-0400-0415-real_values_data.json'
# 
index = 2
plot_single_real_value(file_path, index)
```
